# 1. Thermische & Meteorologische Features
Spalten: mittlere_temperatur_c, heiztage

Best Practice: Die "Grundlast" (Baseload) vs. Heizlast trennen
In der Realität verbraucht ein Gebäude Strom/Gas für Dinge, die immer laufen (Licht, Server, Maschinen = Grundlast) und Energie, die rein temperaturabhängig ist.

- Heizintensität pro Heiztag: Wie viel Leistung/Energie wird wirklich pro mathematischem Heiztag abgerufen?$$\text{Verbrauch pro Heiztag} = \frac{\text{Gesamtverbrauch}}{\text{Heiztage}}

- Temperatur-Nichtlinearität (U-Kurve): Energieverbrauch verhält sich nicht linear zur Temperatur.Bei sehr niedrigen Temperaturen steigt er (Heizung), in einem Wohlfühlbereich (ca. $15^\circ\text{C}$ bis $22^\circ\text{C}$) bleibt er flach, und bei sehr hohen Temperaturen steigt er wieder (Klimatisierung).

Praxis-Tipp für ML-Modell: Erstelle Quadrate der Temperatur (mittlere_temperatur_c * mittlere_temperatur_c), um diese U-Kurve für Algorithmen abbildbar zu machen.

# 2. Kalendarische & Zeitliche Features
Spalten: monat_idx, arbeitstage, feiertage_im_monat

Best Practice: Der "Produktivitäts-Quotient"Ein reiner Monat (z.B. Januar) sagt wenig aus, da er jedes Jahr eine andere Anzahl an echten Werktagen besitzt.

- Netto-Arbeitstage-Verhältnis (Aktivitäts-Index): Welcher Anteil des Monats wurde tatsächlich gearbeitet?$$\text{Aktivitäts-Index} = \frac{\text{arbeitstage}}{\text{arbeitstage} + \text{feiertage_im_monat}}

- Feiertags-Disruption-Faktor: In der Energiewirtschaft brechen Verbräuche an Feiertagen vor allem bei der Industrie massiv ein, während sie in Kommunen/Wohnbereichen manchmal leicht steigen. Das Verhältnis zeigt dem Modell sofort die Sensitivität des Kundentyps.

# 3. Vertragliche & Kapazitäts-Features
Spalte: vertragsleistung_kw

Best Practice: "Peak-to-Load"-Verhältnis (Gleichzeitigkeitsfaktor)
- Auslastungsgrad (Capacity Utilization): Wie nah operiert ein Kundentyp an der Grenze seiner gebuchten Infrastruktur? Wenn ein Kunde 500 kW Vertragsleistung gebucht hat, aber im Schnitt nur umgerechnet 50 kW nutzt, ist er "überdimensioniert". Das ist eine extrem wertvolle betriebswirtschaftliche Erkenntnis für Energieversorger (Netzentgelte!).
